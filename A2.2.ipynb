```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as LA

def sim_game(dimension, rng):
    """Simulate a single game for a matrix of given dimension."""
    M = rng.integers(0, 2, (dimension, dimension))
    detM = LA.det(M)
    return detM

def calculate_zero_det_probability(dimension, sample):
    """Calculate the probability of a zero determinant for a given dimension."""
    rng = np.random.default_rng()
    zero_det_count = sum(sim_game(dimension, rng) == 0 for _ in range(int(sample)))
    return zero_det_count / sample

def main():
    max_dimension = 8  # Maximum matrix dimension to simulate
    sample = 1e5  # Number of simulations per dimension

    # Calculate probabilities for each dimension
    dimensions = range(1, max_dimension + 1)
    probabilities = [calculate_zero_det_probability(dim, sample) for dim in dimensions]

    # Plotting the results
    plt.plot(dimensions, probabilities, marker='o', color='purple')
    plt.xlabel('Matrix Dimension')
    plt.ylabel('Probability of Zero Determinant')
    plt.title('Probability of Zero Determinant as Matrix Dimension Increases')
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
